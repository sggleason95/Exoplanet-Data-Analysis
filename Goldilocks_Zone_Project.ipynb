```python
import pandas as pd
import matplotlib.pyplot as plt

# Load file
df = pd.read_csv('exoplanets.csv')

# Look at the first 5 rows to prove it worked
print(df.head())

# See how many planets were found by each method
print(df['discoverymethod'].value_counts())

habitable_candidates = df[(df['pl_eqt']>200) & (df['pl_eqt']<320)]

print(f"I found {len(habitable_candidates)} candidates in the habitable zone!")
print(habitable_candidates[['pl_name', 'hostname', 'pl_eqt']])






```

```python
from google.colab import data_table

# Turns on the interactive table view
data_table.enable_dataframe_formatter()

# Displays your filtered list
# habitable_candidates.dropna() - Too many columns, displays default view

# Only show the name, star, and temperature columns
habitable_candidates[['pl_name', 'hostname', 'pl_eqt']].dropna()

# Create a histogram of the temperatures
habitable_candidates['pl_eqt'].hist(bins=20, color='skyblue', edgecolor='black')

# Add some labels so we know what we are looking at
import matplotlib.pyplot as plt
plt.title('Temperature Distribution of Habitable Candidates')
plt.xlabel('Temperature (Kelvin)')
plt.ylabel('Number of Planets')

```

```python
kelvin_temp = 300

# Convert Kelvin to Celsius
celsius_temp = kelvin_temp - 273.15

# Convert Celsius to Fahrenheit
fahrenheit_temp = (celsius_temp * 9/5) + 32

print(f"{kelvin_temp} Kelvin is equal to {celsius_temp:.2f} Celsius.")
print(f"{kelvin_temp} Kelvin is equal to {fahrenheit_temp:.2f} Fahrenheit.")
```
